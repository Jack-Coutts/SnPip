```python
import pandas as pd
import gzip
```

```python
with gzip.open("resources_broad_hg38_v0_Mills_and_1000G_gold_standard.indels.hg38.vcf.gz", "rt") as f:
    lines = f.readlines()
    
    chrom_index = [i for i, line in enumerate(lines) if line.strip().startswith("#CHROM")]
    data = lines[chrom_index[0]:]  
    header = data[0].strip().split("\t")
    informations = [d.strip().split("\t") for d in data[1:]]

vcf = pd.DataFrame(informations, columns=header)
```

```python
with open("resources_broad_hg38_v0_Mills_and_1000G_gold_standard.indels.hg38.vcf", "r") as f:
    lines = f.readlines()
    
    chrom_index = [i for i, line in enumerate(lines) if line.strip().startswith("#CHROM")]
    data = lines[chrom_index[0]:]  
    header = data[0].strip().split("\t")
    informations = [d.strip().split("\t") for d in data[1:]]

vcf = pd.DataFrame(informations, columns=header)
```

```python
# collected

def get_vcf_names(vcf_path):
    with gzip.open(vcf_path, "rt") as ifile:
        for line in ifile:
            if line.startswith("#CHROM"):
                vcf_names = [x for x in line.strip().split('\t')]
                break
    ifile.close()
    return vcf_names


names = get_vcf_names('resources_broad_hg38_v0_Mills_and_1000G_gold_standard.indels.hg38.vcf.gz')
vcf = pd.read_csv('resources_broad_hg38_v0_Mills_and_1000G_gold_standard.indels.hg38.vcf.gz', \
            compression='gzip', comment='#', chunksize=10000, delim_whitespace=True, header=None, names=names)
```
